```python
"""
Create a match table for osprey photos to go with nest locations.
"""

import arcpy
import os

# Nest locations table
nest_locations = "Survey/Osprey_Nest_Location"

# Extract site id from the photo name
# Match that to the nest location table. 

input = r"F:\GIS\DB_CONNECT\Collection.sde\SDE.Survey\SDE.Osprey_Nest_Location"
inputField = "Nest_ID"
matchTable = r"matchtable.csv"
matchField = "Nest_ID"
pathField = "Picture" 
picFolder = r'F:\GIS\GIS_DATA\Wildlife\Osprey\Osprey_Survey\Survey_Photos\2023'

def get_chars_after(main_string, defined_string, num_chars=5):
    # Find the starting index of the defined string
    start_index = main_string.find(defined_string)
    
    # If the defined string is not found, return an empty string
    if start_index == -1:
        return ""
    
    # Calculate the starting index of the substring we want to extract
    start_index += len(defined_string)
    
    # Extract the substring with the defined number of characters
    result = main_string[start_index:start_index + num_chars]
    
    return result

# Create the CSV file and write the header row
with open(matchTable, "w") as csv_file:
    # Write the header row
    csv_file.write(f"{matchField},{pathField}\n")
    
    # Iterate through each picture in the directory and write a row to the table
    for file in os.listdir(picFolder):
        if file.endswith(".jpg"):
            # Extract the ParcelID by removing the '.jpg' extension from the file name
            parcel_id = get_chars_after(file, "Osprey_Nest_", 5)
            # Write the ParcelID and Picture (file name) to the CSV file
            invalid_photo_names = ['Photo','New_2', 'Old_n']
            if parcel_id not in invalid_photo_names:
                csv_file.write(f"{parcel_id},{file}\n")
```

```python

# The input feature class must first have GDB attachments enabled
#arcpy.management.EnableAttachments(input)

# Use the match table with the Add Attachments tool
arcpy.management.AddAttachments(input, inputField, matchTable, matchField, 
                                pathField, picFolder)
```
